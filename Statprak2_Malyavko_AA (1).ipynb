```python
import matplotlib.pyplot as plt
X = np.random.rand(1000, 2)
def kmeans(X, k, itr, thresh):
    centroids = X[:k]
    for i in range(itr):
        labels = np.argmin(np.linalg.norm(X - centroids[:, None], axis=2), axis=0)
        newcentroids = np.array([X[labels==j, :].mean(axis=0) for j in range(k)])
        dist = ((newcentroids - centroids)**2).sum()
        if (dist <= thresh) or (newcentroids == centroids).all():
            break
        else:
            centroids = newcentroids
    return(centroids, dist, labels, i)

centroids, dist, labels, i = kmeans(X, 10, 100, 1e-05)       
plt.scatter(x=X[:, 0], y=X[:, 1], s=20, c=labels)
plt.scatter(x=centroids[:, 0], y=centroids[:, 1], s=100, marker=(9, 2))
plt.title(f" {i} K-Means Iterations: Predicted Classes")
plt.show()
```
